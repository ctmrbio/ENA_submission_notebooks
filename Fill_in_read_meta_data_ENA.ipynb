```python
import pandas as pd 
from pathlib import Path
```

```python
dysp_md5_sums = pd.read_csv("/ceph/projects/031_VaMiGyn/ENA_upload/host_removed/md5lists_240409/md5list.txt",header=None,sep=" ")
dysp_md5_sums["Samples"] = dysp_md5_sums[2].str[:-8]
dysp_md5_sums["forward_reads"] = dysp_md5_sums[2][dysp_md5_sums[2].str.endswith('_1.fq.gz')]
dysp_md5_sums["forward_reads_md5sums"] = dysp_md5_sums[0][dysp_md5_sums[2].str.endswith('_1.fq.gz')]

dysp_md5_sums["reverse_reads"] = dysp_md5_sums[2][dysp_md5_sums[2].str.endswith('_2.fq.gz')]
dysp_md5_sums["reverse_reads_md5sums"] = dysp_md5_sums[0][dysp_md5_sums[2].str.endswith('_2.fq.gz')]
dysp_md5_sums["SampleID"] = dysp_md5_sums["Samples"]


dysp_md5_sums.head()
```

```python
all_samples_meta_data = pd.read_csv("/ceph/projects/031_VaMiGyn/gustav.ahlstrom/ENA_upload/240407_VMG_list_forENA.csv",header=0,sep=";")


all_samples_meta_data.index = all_samples_meta_data["SampleID"]


all_samples_meta_data

```

```python
checklist_filled = pd.DataFrame(index=range(int(dysp_md5_sums.shape[0]/2)),columns=range(12))
checklist_filled.columns = ['sample', 'study', 'instrument_model', 'library_name', 'library_source',
       'library_selection', 'library_strategy', 'library_layout',
       'forward_file_name', 'forward_file_md5', 'reverse_file_name',
       'reverse_file_md5']

unique_samples = dysp_md5_sums["Samples"].drop_duplicates()
unique_samples.index = range(len(unique_samples))
checklist_filled["sample"] = unique_samples
checklist_filled["study"] = "PRJEB74625"
checklist_filled["instrument_model"] = "DNBSEQ-T7"
checklist_filled["library_source"] = "METAGENOMIC"
checklist_filled["library_selection"] = "PCR"
checklist_filled["library_strategy"] = "WGA"
checklist_filled["library_layout"] = "PAIRED"
checklist_filled["library_name"] = unique_samples+" WGS data"

unique_forward_reads = dysp_md5_sums["forward_reads"].dropna()
unique_forward_reads.index = range(len(unique_forward_reads))
unq_fwd_rds_md5 = dysp_md5_sums["forward_reads_md5sums"].dropna()
unq_fwd_rds_md5.index = range(len(unq_fwd_rds_md5))
checklist_filled["forward_file_name"] = unique_forward_reads
checklist_filled["forward_file_md5"] = unq_fwd_rds_md5

unq_rev_reads = dysp_md5_sums["reverse_reads"].dropna()
unq_rev_reads.index = range(len(unq_rev_reads))
unq_rev_rds_md5 = dysp_md5_sums["reverse_reads_md5sums"].dropna()
unq_rev_rds_md5.index = range(len(unq_rev_rds_md5))
checklist_filled["reverse_file_name"] = unq_rev_reads
checklist_filled["reverse_file_md5"] = unq_rev_rds_md5
```

```python
checklist_filled
```

```python
new_row_name = pd.DataFrame({'sample':['sample'],
                             'study':['study'],
                             'instrument_model':['instrument_model'],
                             'library_name':['library_name'], 
                             'library_source':['library_source'],
                             'library_selection':['library_selection'], 
                             'library_strategy':['library_strategy'], 
                             'library_layout':['library_layout'],
                             'forward_file_name':['forward_file_name'], 
                             'forward_file_md5':['forward_file_md5'], 
                             'reverse_file_name':['reverse_file_name'],
                             'reverse_file_md5':['reverse_file_md5']})
column_names_proper = ["FileType","fastq","Read submission file type","","","","","","","","",""]



checklist_filled = pd.concat([pd.DataFrame(new_row_name, index=[0]), checklist_filled])
checklist_filled.columns = column_names_proper
checklist_filled
```

```python
#Save data now
checklist_filled.to_csv('/ceph/projects/031_VaMiGyn/gustav.ahlstrom/ENA_upload/reads_meta_data_filled_in_VaMiGyn.tsv',sep="\t",index=False, header=True)

```
