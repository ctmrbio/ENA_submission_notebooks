```python
import pandas as pd 
from pathlib import Path
import shutil
```

```python
all_samples = pd.read_csv("../../../ENA_upload/host_removed/samples_all/all_samples.txt",header=0,sep=";")
all_samples.head()
```

```python
all_samples["FC_ID"] = all_samples["Samples_s3"].str.split("__", expand=True)[2]
all_samples = all_samples.set_index("FC_ID")
all_samples.head()
type(all_samples.index[1])
```

```python
samples_upload = pd.read_csv("../240407_VMG_list_forENA.csv",header=0,sep=";")
samples_upload.head()
type(samples_upload["Flowcell_L1_bc"][1])
```

```python
Path("../../ENA_upload/host_removed/symlink_ENA_April24_upload/").mkdir(parents=True,exist_ok=True)
```

```python
for idx,row in samples_upload.iterrows():
    sample_id = row["SampleID"]
    fc_id = row["Flowcell_L1_bc"]
    print(sample_id)
    print(fc_id)
    read_1 = all_samples.loc[fc_id].iloc[0]["Samples_s3"]
    #read_1 = all_samples.loc[fc_id]
    print(read_1)
    read_2 = all_samples.loc[fc_id].iloc[1]["Samples_s3"]
    print(read_2)
    read_1_abs = Path(f"/ceph/projects/031_VaMiGyn/ENA_upload/host_removed/samples_all/{read_1}").resolve()
    read_2_abs = Path(f"/ceph/projects/031_VaMiGyn/ENA_upload/host_removed/samples_all/{read_2}").resolve()
    
    ena_1 = Path(f"/ceph/projects/031_VaMiGyn/ENA_upload/host_removed/symlink_ENA_April24_upload/{sample_id}_1.fq.gz") 
    ena_2 = Path(f"/ceph/projects/031_VaMiGyn/ENA_upload/host_removed/symlink_ENA_April24_upload/{sample_id}_2.fq.gz") 

    try: 
        ena_1.hardlink_to(read_1_abs)
        ena_2.hardlink_to(read_2_abs)
        print(sample_id)
    except FileExistsError as e:
        pass 
```
